```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

##### Basic Graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.plot(x,y)  #plots a line in a graph. takes 1d array as arguments.
plt.title('My graph')  #used to give title to the plot
plt.xlabel('x-axis')   #used to assign label to the x-axis of the graph
plt.ylabel('y-axis')   #used to assign label to the y-axis of the graph
plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.plot(x,y, label='2x',color='red',linewidth='2',marker='.',markersize='10')  #Visit the official matplotlib docs to learn more about the functions
plt.title('My graph',fontdict={'fontname':'Comic Sans MS','fontsize':'20'})  #fontdict is used to modify the font of the plot
plt.xlabel('x-axis')   #used to assign label to the x-axis of the graph
plt.ylabel('y-axis')   #used to assign label to the y-axis of the graph

plt.xticks([0,1,2,3,4])   #Get or set the current tick locations and labels of the x-axis.
plt.yticks([0,2,4,6,8,10]) #Get or set the current tick locations and labels of the y-axis.

plt.legend() #used to add a legend for reference
plt.show()
```

##### Docs: https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html#functions

```python
#use shorthand notation
#format= '[color][marker][line]'

plt.plot(x,y,'r^--', label='2x')  #Visit the official matplotlib docs to learn more about the functions
plt.title('My graph',fontdict={'fontname':'Comic Sans MS','fontsize':'20'})  #fontdict is used to modify the font of the plot
plt.xlabel('x-axis')   #used to assign label to the x-axis of the graph
plt.ylabel('y-axis')   #used to assign label to the y-axis of the graph

plt.xticks([0,1,2,3,4])   #Get or set the current tick locations and labels of the x-axis.
plt.yticks([0,2,4,6,8,10]) #Get or set the current tick locations and labels of the y-axis.

plt.legend() #used to add a legend for reference
plt.show()
```

##### Adding a second line

```python
#use shorthand notation
#format= '[color][marker][line]'

plt.plot(x,y,'r^--', label='2x')  #Visit the official matplotlib docs to learn more about the functions

#second line
x2=np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,label='x^2')

plt.title('My graph',fontdict={'fontname':'Comic Sans MS','fontsize':'20'})  #fontdict is used to modify the font of the plot
plt.xlabel('x-axis')   #used to assign label to the x-axis of the graph
plt.ylabel('y-axis')   #used to assign label to the y-axis of the graph

plt.xticks([0,1,2,3,4])   #Get or set the current tick locations and labels of the x-axis.
#plt.yticks([0,2,4,6,8,10]) #Get or set the current tick locations and labels of the y-axis.

plt.legend() #used to add a legend for reference
plt.show()
```

```python
#use shorthand notation
#format= '[color][marker][line]'

#resize your graph
plt.figure(figsize=(5,3),dpi=100)

plt.plot(x,y,'r^--', label='2x')  #Visit the official matplotlib docs to learn more about the functions

#second line
x2=np.arange(0,4.5,0.5)
#plot part of the graph as a line
plt.plot(x2[:5],x2[:5]**2,'b',label='x^2')

#plot the remainder of the graph
plt.plot(x2[4:],x2[4:]**2,'b.--')

plt.title('My graph',fontdict={'fontname':'Comic Sans MS','fontsize':'20'})  #fontdict is used to modify the font of the plot
plt.xlabel('x-axis')   #used to assign label to the x-axis of the graph
plt.ylabel('y-axis')   #used to assign label to the y-axis of the graph

plt.xticks([0,1,2,3,4])   #Get or set the current tick locations and labels of the x-axis.
#plt.yticks([0,2,4,6,8,10]) #Get or set the current tick locations and labels of the y-axis.

plt.legend() #used to add a legend for reference

plt.savefig('graph.png',dpi=300)  #used to save the graph as a file
plt.show()
```

#### Bar Chart

```python
labels=['A','B','C']
values=[1,4,2]
bars=plt.bar(labels,values)

patterns=['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))  #set_hatch function is used to set bar patterns
plt.show()
```
